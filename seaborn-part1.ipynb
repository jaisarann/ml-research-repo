```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
df.dtypes
```

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

### Join Plot - Uni / Bivariate Analysis

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='scatter')
```

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='hex')
```

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='reg')
```

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='resid')
```

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='kde')
```

### Pair plot - more than two features

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue='sex')
```

### Distributed Plot

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False)
```
