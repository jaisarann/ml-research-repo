```python
from numpy import asarray
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder, LabelEncoder, LabelBinarizer
```

```python
data = asarray([['red'], ['green'], ['blue']])
data, type(data), data.shape
```

```python
help(LabelBinarizer)
```

```python
LB = LabelBinarizer()
LB.fit_transform(['red', 'green', 'blue'])
```

```python
LE = LabelEncoder()
le_data = LE.fit_transform(['red', 'green', 'blue'])
le_data
```

```python
OE = OrdinalEncoder()
t_data = OE.fit_transform(data)
t_data
```

```python
OHE = OneHotEncoder(categories=[sparse=False, drop='first')
ohe_data = OHE.fit_transform(data)
ohe_data
```
