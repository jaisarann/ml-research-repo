```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,10)
y = np.arange(11,21)
```

### Scatter

```python
plt.scatter(x, y, c='r')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('scatter plot')
```

```python
plt.plot(x,y,'r*--')
```

```python
y1 = x
```

```python
plt.plot(x,y1)
```

### Subplot

```python
plt.subplot(2,2,1)
plt.plot(x,y, 'g')
plt.subplot(2,2,2)
plt.plot(x,y, 'r')
plt.subplot(2,2,3)
plt.plot(x,y, 'y')
```

## Plot

```python
x2 = np.arange(1,11)
y2 =50 * x2 + 10
plt.plot(x2, y2, 'r--')
```

```python
x3 = np.arange(0,10,.1)
y_sin = np.sin(x3)
plt.plot(x3, y_sin, 'r')
```

```python
x4 = np.arange(0,10,.1)
y_cos = np.cos(x3)
plt.plot(x4, y_cos, 'r')
```
